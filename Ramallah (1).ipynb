# üîç Step 3: Exploratory Data Analysis (EDA)

```python
print("üì¶ Number of unique items:", df['Food Menu(Code)'].nunique())
print("üßæ Total quantity sold:", df['Quantity'].sum())
print("üí∞ Total revenue:", df['Sales Value'].sum())
print("üíµ Average selling price overall:", df['Ave Selling Price'].mean())
```

```python
top_items = df.groupby('Food Menu(Code)')['Quantity'].sum().sort_values(ascending=False).head(10)
print(top_items)
```

```python
import matplotlib.pyplot as plt
top_items.plot(kind='bar', figsize=(10,5), title='Top 10 Selling Items (By Quantity)', color='skyblue')
plt.ylabel('Quantity Sold')
plt.xlabel('Item')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("Ramallah4.png")
plt.show()
```

```python
top_revenue = df.groupby('Food Menu(Code)')['Sales Value'].sum().sort_values(ascending=False).head(10)
print(top_revenue)
```

```python
top_revenue.plot(kind='bar', figsize=(10,5), title='Top 10 Revenue-Generating Items', color='gold')
plt.ylabel('Total Sales Value')
plt.xlabel('Item')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('top_revenue.png')
plt.show()
```

```python
cat_revenue = df.groupby('Category Name')['Sales Value'].sum().sort_values(ascending=False)
print(cat_revenue)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
cat_revenue.plot(kind='bar', color='teal')

plt.title('Total Revenue by Category', fontsize=16, weight='bold')
plt.xlabel('Category Name', fontsize=13)
plt.ylabel('Revenue (EGP)', fontsize=13)
plt.xticks(rotation=30, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig('top_revenue(Cat).png')
plt.show()
```

```python
cat_quantity = df.groupby('Category Name')['Quantity'].sum().sort_values(ascending=False)
print(cat_quantity)
```

```python
plt.figure(figsize=(12, 12))
cat_revenue.plot(kind='pie', autopct='%1.1f%%', startangle=140, shadow=True, cmap='Set3')
plt.title("Total Quntity by Category")
plt.ylabel('')
plt.savefig('top_sales(Cat).png')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Order categories by median price
ordered_categories = df.groupby('Category Name')['Ave Selling Price'].median().sort_values(ascending=False).index

plt.figure(figsize=(14, 6))
sns.set(style="whitegrid")

# Improved boxplot
sns.boxplot(
    data=df,
    x='Category Name',
    y='Ave Selling Price',
    order=ordered_categories,
    palette='Set2'
)

plt.title('Average Selling Price Distribution by Category', fontsize=16, weight='bold')
plt.xlabel('Category Name', fontsize=13)
plt.ylabel('Average Selling Price', fontsize=13)
plt.xticks(rotation=30, ha='right', fontsize=10)
plt.tight_layout()
plt.grid(True, axis='y', linestyle='--', alpha=0.6)

plt.show()
```

```python
category_revenue = df.groupby('Category Name')['Sales Value'].sum().sort_values(ascending=False)
category_quantity = df.groupby('Category Name')['Quantity'].sum().sort_values(ascending=False)
```

```python
category_revenue
```

```python
category_quantity
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Combine revenue and quantity into one DataFrame
category_stats = pd.DataFrame({
    'Revenue': category_revenue,
    'Quantity': category_quantity
})

# Normalize the values for better visual comparison (optional but recommended)
category_stats_norm = category_stats.copy()
category_stats_norm['Revenue'] = category_stats['Revenue'] / category_stats['Revenue'].max()
category_stats_norm['Quantity'] = category_stats['Quantity'] / category_stats['Quantity'].max()

# Plot
category_stats_norm.plot(kind='bar', figsize=(14, 6), color=['seagreen', 'steelblue'])

plt.title('Normalized Revenue vs Quantity by Category', fontsize=16, weight='bold')
plt.xlabel('Category', fontsize=13)
plt.ylabel('Normalized Value (0‚Äì1)', fontsize=13)
plt.xticks(rotation=30, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.legend(title='Metric')
plt.tight_layout()
plt.savefig('revnue vs qui.png')

plt.show()
```
